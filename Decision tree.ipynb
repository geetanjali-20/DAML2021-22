```python
X=df.drop('salary_more_then_100k',axis='columns')
y=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
L_Company=LabelEncoder()
L_job=LabelEncoder()
L_degree=LabelEncoder()
```

```python
X['c']=L_Company.fit_transform(X['company'])
X['j']=L_job.fit_transform(X['job'])
X['d']=L_degree.fit_transform(X['degree'])
```

```python
df1=X.drop(['company','job','degree'],axis=1)
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df1,y,random_state=42,test_size=0.25)
```

```python
from sklearn import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```
